# Linear Algebra

##### <u>Multiplication</u>

```python
# For multiplication, rows of 1st = columns of 2nd
a = np.ones((2,3))
print(a)

b = np.full((3,2), 2)
print(b)

np.matmul(a,b)
```

##### <u>Find the determinant</u>

```python
# ID matrix has determinant of 1
c = np.identity(3)
np.linalg.det(c)
```

# Statistics

```python
stats = np.array([[1,2,3],[4,5,6]])
stats
```

##### Minimum

```python
np.min(stats)
```

##### Minimum based on row

```python
np.min(stats, axis=1) # 0 row (side) # 1 column (down)
```

##### Maximum

```python
np.max(stats)
```

##### Sum

```python
np.sum(stats, axis=0)
```
